```python
# Existense of this notebook is to show that the images are read with their outlines which in the long run will cause problems related with the edge detection 
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
img_original = cv2.imread('images/fezandipiti.png',cv2.IMREAD_UNCHANGED) 

img = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)
img_Canny = cv2.Canny(img, 200, 250)
```

```python
# This cell properly reads the image with alpha channel then changes the transparent background with the white background
mask = img_original[:,:,3] == 0                                          
img_original[mask] = [255, 255, 255, 255]
```

```python
img2 = cv2.cvtColor(img_original, cv2.COLOR_BGRA2RGB)
img_Canny2 = cv2.Canny(img2, 200, 250)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].imshow(img)
axes[0, 0].set_title("Original Image")
axes[0, 0].axis("off")

axes[0, 1].imshow(img2)
axes[0, 1].set_title("Converted Image (img2)")
axes[0, 1].axis("off")

axes[1, 0].imshow(img_Canny, cmap="gray")
axes[1, 0].set_title("Canny Edge Detection (img_Canny)")
axes[1, 0].axis("off")

axes[1, 1].imshow(img_Canny2, cmap="gray")
axes[1, 1].set_title("Canny Edge Detection (img_Canny2)")
axes[1, 1].axis("off")
```
